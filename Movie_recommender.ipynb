```python
import numpy as np
import pandas as pd
```

```python
movies=pd.read_csv("/content/tmdb_5000_movies.csv")
credits=pd.read_csv("/content/tmdb_5000_credits.csv")
```

```python
# Selecting important columns.
# genres,id,keywords,overview,title,cast,crew.
movies=movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head(1)
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

```python
# For duplicate values.
movies.duplicated().sum()
```

```python
# Working on genre column.
movies.iloc[0].genres
```

```python
#Convert string of list into list.
import ast
```

```python
# I want in ['Action','Adventure']
def convert(obj):
  l=[]
  for i in ast.literal_eval(obj):
    l.append(i['name'])
  return l
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head(1)
```

```python
# Similar for Keywords.
movies['keywords']=movies['keywords'].apply(convert)
movies.head(1)
```

```python
# For cast I want top 3 actors.
def convert(obj):
  l=[]
  count=0
  for i in ast.literal_eval(obj):
    if count !=3:
      l.append(i['name'])
      count+=1
    else:
      break
  return l
```

```python
movies['cast']=movies['cast'].apply(convert)
```

```python
movies.head(1)
```

```python
# For crew I want only director name.
def fetch_director(obj):
  l=[]

  for i in ast.literal_eval(obj):
    if i['job']=='Director':
      l.append(i['name'])
      break
  return l
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
movies.head(1)
```

```python
# For overview convert to list such that it is easily concat with my others.
movies['overview']=movies['overview'].apply(lambda x:x.split())
movies.head(1)
```

```python
# I want to remove spaces btw names bcz to remove confusion like John Cena or John Snow
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head(2)
```

```python
# For creating tags i merge genres,keywords,top cast and director
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
movies.head(1)
```

```python
# Drop remaining one.
new_df=movies[['movie_id','title','tags']]
new_df.head(2)
```

```python
# Convert tags into string.
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
new_df.head(2)
```

```python
# convert tags into lower()
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
# Using Text vectorization. Convert each movie into vectors, by Bag-of-Words.
# Firstly remove stopwords.
# Apply Stemming.
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
# I want to calculate cosine similarity.
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similar=cosine_similarity(vectors)
```

```python
similar.shape
```

```python
def recommened(movie):
  movie_index=new_df[new_df['title']==movie].index[0]
  distances=similar[movie_index]
  # But before sorting i want to catch index of sorting movie with the help of enumerate function.
  movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
  for i in movies_list:
    print(new_df.iloc[i[0]].title)
```

```python
recommened("Avatar")
```

```python
recommened("Batman Begins")
```
